```python
df4 = pd.DataFrame({ 'key': ['A4','A5','A6','A7'],
    'E': ['A4','A5','A6','A7'],
    'F': ['B4','B5','B6','B7'],
    'G': ['C4','C5','C6','C7'],
    'H': ['D4','D5','D6','D7']
},
index=[4,5,6,7])
```

```python
#merge operation
pd_mrg_inr = pd.merge(df1,df4,how='inner',on='key')
pd_mrg_inr
```

```python
df5 = pd.DataFrame({ 'key1': ['A4','A5','A6','A7'],
    'E': ['A4','A5','A6','A7'],
    'F': ['B4','B5','B6','B7'],
    'G': ['C4','C5','C6','C7'],
    'H': ['D4','D5','D6','D7']
},
index=[0,1,2,7])
```

```python
df6 = pd.DataFrame({ 'key2': ['A4','A5','A10','A12'],
    'A': ['A0','A1','A2','A3'],
    'B': ['B0','B1','B2','B3'],
    'C': ['C0','C1','C2','C3'],
    'D': ['D0','D1','D2','D3']
},
index=[0,1,2,3])
```

```python
#merge operation if key colmn name is different
pd_mrg_inr2 = pd.merge(df5,df6,how='inner',left_on='key1',right_on='key2')
pd_mrg_inr2
```

```python
#join will alaways join based on indexes
df5.join(df6)
```

```python
pd_mrg_outer = pd.merge(df5,df6,how='outer',left_on='key1',right_on='key2')
pd_mrg_outer
```
